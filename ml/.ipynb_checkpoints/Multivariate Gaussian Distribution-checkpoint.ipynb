```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
np.random.seed(0)
n_samples = 1000
heights = np.random.normal(loc=170, scale=10, size=n_samples)
weights = np.random.normal(loc=70, scale=15, size=n_samples)
data = np.vstack((heights, weights)).T
```

```python
# Calculate mean and covariance
mean = np.mean(data, axis=0)
covariance = np.cov(data, rowvar=False)
print(f"Mean (Height, Weight): {mean}")
print(f"Covariance Matrix:\n{covariance}")
```

```python
x = np.linspace(140, 200, 100)
y = np.linspace(30, 120, 100) 
X, Y = np.meshgrid(x, y)
```

```python
pos = np.dstack((X, Y))
rv = multivariate_normal(mean, covariance)
Z = rv.pdf(pos)
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x=heights, y=weights, color='blue', alpha=0.6, label='Data Points')
plt.contour(X, Y, Z, levels=10, cmap='viridis')
plt.title('Multivariate Gaussian Distribution of Height and Weight')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.show()
```
