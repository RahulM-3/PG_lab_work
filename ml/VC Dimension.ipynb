```python
import itertools
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import SGDClassifier
from sklearn.datasets import make_classification
```

```python
def is_shattered(points, labels):
    if(len(set(labels))) <= 1:
        return True
    model = SGDClassifier()
    model.fit(points, labels)
    return int(model.score(points, labels))
```

```python
def calculate_vc_dimension(points, all_labels): 
    for i in range(1, n + 1):
        for labels in all_labels:
            if not is_shattered(points[:i], labels[:i]):
                return i - 1
    return n
```

```python
n = 10
points, _ = make_classification(n_samples=n, n_features=2, 
                            n_informative=2, n_redundant=0, 
                            n_clusters_per_class=1)
all_labels = list(itertools.product([0, 1], repeat=n))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(points[:, 0], points[:, 1], cmap='bwr', edgecolors='k', s=50)
plt.title("Dataset Points")
plt.xlabel("X1")
plt.ylabel("X2")
plt.grid(True)
plt.show()
```

```python
vc_dim = calculate_vc_dimension(points, all_labels)
print("VC Dimension of the given hypothesis set:", vc_dim)
```
