```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
def frequentist_mean(data):
    mean = np.mean(data)
    plt.figure(figsize=(10, 6))
    sns.histplot(data, bins=15, kde=True, stat='density', color='skyblue', label='Data Distribution')
    plt.axvline(mean, color='red', linestyle='--', label=f'Frequentist Mean: {mean:.2f}')
    plt.title('Frequentist Approach to Estimating the Mean of Alcohol Content in Wine')
    plt.xlabel('Alcohol Content (%)')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    return mean
```

```python
def bayesian_mean(data):
    prior_mean = 10
    prior_variance = 5

    n = len(data)
    sample_mean = np.mean(data)
    sample_variance = np.var(data, ddof=1)

    posterior_variance = 1 / (1/prior_variance + n/sample_variance)
    posterior_mean = posterior_variance * (prior_mean/prior_variance + n*sample_mean/sample_variance)

    x = np.linspace(8, 15, 26)
    prior_dist = norm.pdf(x, loc=prior_mean, scale=np.sqrt(prior_variance))
    likelihood_dist = norm.pdf(x, loc=sample_mean, scale=np.sqrt(sample_variance/n))
    posterior_dist = norm.pdf(x, loc=posterior_mean, scale=np.sqrt(posterior_variance))

    plt.figure(figsize=(10, 6))
    sns.histplot(data, bins=15, kde=True, stat='density', color='skyblue', label='Data Distribution')
    plt.plot(x, prior_dist, color='green', label='Prior Distribution', alpha=0.6)
    plt.plot(x, likelihood_dist, color='orange', label='Likelihood Distribution', alpha=0.6)
    plt.plot(x, posterior_dist, color='purple', label='Posterior Distribution', alpha=0.8)
    plt.axvline(posterior_mean, color='purple', linestyle='--', label=f'Bayesian Posterior Mean: {posterior_mean:.2f}')
    plt.title('Bayesian Approach to Estimating the Mean of Alcohol Content in Wine')
    plt.xlabel('Alcohol Content (%)')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

    return posterior_mean
```

```python
data = pd.read_csv("wine.csv", sep=';')
data.head()
```

```python
alcohol_data = data['alcohol']
alcohol_data.head()
```

```python
frequentist_mean(alcohol_data)
```

```python
bayesian_mean(alcohol_data)
```
