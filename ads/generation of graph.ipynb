```python
import networkx as nx
```

```python
class graph:
    def __init__(self, directional=False):
        self.nodes = {}
        self.directional = directional
        self.G = None
        
    def add_node(self, node, parent=None):
        if(parent==None):
            self.nodes[node] = []
        else:
            if(parent not in self.nodes):
                self.nodes[parent] = [node]
            else:
                self.nodes[parent].append(node)

    def generate_graph(self):
        if(self.directional):
            self.G = nx.DiGraph(self.nodes)
        else:
            self.G = nx.Graph(self.nodes)
        return self.G
    
    def display(self):
        print("Nodes:", self.G.nodes())
        print("Edges:", self.G.edges())

    def visualize(self, node_color):
        if(self.directional):
            pos = nx.spring_layout(self.G, seed = 50)
        else:
            pos = None
        nx.draw(self.G, pos = pos, ax = None, with_labels = True, font_size = 20, node_size = 2000, node_color = node_color)
        
```

```python
G = graph()
G.add_node('a')
G.add_node('b', 'a')
G.add_node('g', 'a')
G.add_node('d', 'a')
G.add_node('a', 'b')
G.add_node('g', 'b')
G.add_node('c', 'b')
G.add_node('e', 'b')


print(G.nodes)
G.generate_graph()
G.display()
G.visualize(node_color="green")
```

```python
G = graph(directional=True)
G.add_node('a')
G.add_node('b', 'a')
G.add_node('g', 'a')
G.add_node('d', 'a')
G.add_node('a', 'b')
G.add_node('g', 'b')
G.add_node('c', 'b')
G.add_node('e', 'b')

G.generate_graph()
G.display()
G.visualize(node_color="green")
```
